#### Load libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Load the data

```python
df = pd.read_csv('oversampled.csv')
```

```python
good = df[df.target == 0].drop('target', axis = 1)
resonator = df[df.target == 1].drop('target', axis = 1)
centermass = df[df.target == 2].drop('target', axis = 1)
diaphragm = df[df.target == 3].drop('target', axis = 1)
```

```python
times = np.array(good.columns[:-1])
time = np.array([float(col) for col in times])
```

```python
amplitude_nums = np.append([1, 2, 3, 4, 5], np.random.randint(0, 1000, 10))
```

### Apply Fourier

```python
def generate_fourier(rows, cols, height, width, data, time, amplitude_nums, title):
    amplitudes = []

    count = 0
    
    for i in amplitude_nums:
        amplitude = np.array(data.iloc[i])
        amplitudes.append(amplitude)

    fig, axs = plt.subplots(rows, cols)
    fig.set_figheight(height)
    fig.set_figwidth(width)
    fig.suptitle('Fourier Transform of ' + title)

    n = len(time)
    dt = time[1] - time[0]

    for amplitude, ax in zip(amplitudes, axs.flat):
        freq = np.fft.fftfreq(n, dt)
        amplitude_fft = np.fft.fft(amplitude)[:-1]

        ax.plot(freq, np.abs(amplitude_fft))
        ax.set_title('Amplitdue ' + str(amplitude_nums[count]))
        ax.set_xlabel('Frequency (Hz)')
        ax.set_ylabel('Amplitude')
        count = count + 1

    plt.show()
```

#### Good Horn

```python
generate_fourier(3, 5, 20, 20, good, time, amplitude_nums, 'Good Horn')
```

#### Resonator Damage

```python
generate_fourier(3, 5, 20, 20, resonator, time, amplitude_nums, 'Resonator Damage')
```

#### Centermass Damage

```python
generate_fourier(3, 5, 20, 20, centermass, time, amplitude_nums, 'Centermass Damage')
```

#### Diaphragm Damage

```python
generate_fourier(3, 5, 20, 20, diaphragm, time, amplitude_nums, 'Diaphragm Damage')
```
