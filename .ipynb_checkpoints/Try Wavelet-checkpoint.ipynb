#### Load libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import cwt, morlet
```

#### Load dataset

```python
df = pd.read_csv('oversampled.csv')
```

```python
good = df[df.target == 0].drop('target', axis = 1)
resonator = df[df.target == 1].drop('target', axis = 1)
centermass = df[df.target == 2].drop('target', axis = 1)
diaphragm = df[df.target == 3].drop('target', axis = 1)
```

```python
times = np.array(good.columns[:-1])
time = np.array([float(col) for col in times])
```

```python
amplitude_nums = np.append([1, 2, 3, 4, 5], np.random.randint(0, 1000, 10))
```

### Apply Wavelet

#### Generating function

```python
def generate_morlet_wavelet(rows, cols, height, width, data, amplitude_nums, time, title):
    amplitudes = []
    
    for i in amplitude_nums:
        amplitude = np.array(data.iloc[i])
        amplitudes.append(amplitude)

    fig, axs = plt.subplots(rows, cols)
    fig.set_figheight(height)
    fig.set_figwidth(width)
    fig.suptitle('Morlet Wavelet Transform of ' + title)

    count = 0
    
    # Parameters
    sampling_rate = 1 / (time[1] - time[0])
    
    # range of frequencies - can be adjusted later
    wavelet_freqs = np.arange(1, 50, 1) 
    
    for amplitude, ax in zip(amplitudes, axs.flat):
        
        # Morlet wavelet transform
        transformed_data = cwt(amplitude, morlet, wavelet_freqs, sampling_rate)
        
        # Plot the results
        im = ax.imshow(np.abs(transformed_data), aspect='auto', extent=[time.min(), time.max(), wavelet_freqs.min(), wavelet_freqs.max()], cmap='jet')
        
        # Add title, x-label, y-label
        ax.set_title('Amplitude ' + str(amplitude_nums[count]))
        ax.set_xlabel('Time [sec]')
        ax.set_ylabel('Frequency [Hz]')
        count = count + 1
        
    # Add colorbar
    fig.colorbar(im, ax=axs.ravel().tolist(), label='Power/Frequency (dB/Hz)')

    plt.show()
```

#### Good Horn

```python
generate_morlet_wavelet(3, 5, 20, 20, good, amplitude_nums, time, 'Good Horn')
```

#### Resonator Damage

```python
generate_morlet_wavelet(3, 5, 20, 20, resonator, amplitude_nums, time, 'Resonator Damage')
```

#### Centermass Damage

```python
generate_morlet_wavelet(3, 5, 20, 20, centermass, amplitude_nums, time, 'Centermass Damage')
```

#### Diaphragm Damage

```python
generate_morlet_wavelet(3, 5, 20, 20, diaphragm, amplitude_nums, time, 'Diaphragm Damage')
```

## Generate Images

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import cwt, morlet
from tqdm import tqdm

# Load the data
df = pd.read_csv('oversampled.csv')

# Separate the data based on the target
good = df[df.target == 0].drop('target', axis = 1)
resonator = df[df.target == 1].drop('target', axis = 1)
centermass = df[df.target == 2].drop('target', axis = 1)
diaphragm = df[df.target == 3].drop('target', axis = 1)

# Convert time to float
times = np.array(good.columns[:-1])
time = np.array([float(col) for col in times])

# Create directories if not exist
directories = ['good', 'resonator', 'centermass', 'diaphragm']
for directory in directories:
    if not os.path.exists(directory):
        os.makedirs(directory)

def generate_and_save_images(data, directory):
    for i in tqdm(range(data.shape[0]), desc=f"Generating images for {directory}"):
        fig, ax = plt.subplots()
        amplitude = np.array(data.iloc[i])
        sampling_rate = 1 / (time[1] - time[0])
        wavelet_freqs = np.arange(1, 50, 1)
        transformed_data = cwt(amplitude, morlet, wavelet_freqs, sampling_rate)
        im = ax.imshow(np.abs(transformed_data), aspect='auto', extent=[time.min(), time.max(), wavelet_freqs.min(), wavelet_freqs.max()], cmap='jet')
        # ax.set_title('Amplitude ' + str(i))
        ax.set_xlabel('Time [sec]')
        ax.set_ylabel('Frequency [Hz]')
        fig.colorbar(im, ax=ax, label='Power/Frequency (dB/Hz)')
        plt.savefig(f"{directory}/Sample {i}.png")
        plt.close(fig)

# Generate and save images
generate_and_save_images(good, 'good')
generate_and_save_images(resonator, 'resonator')
generate_and_save_images(centermass, 'centermass')
generate_and_save_images(diaphragm, 'diaphragm')
```
