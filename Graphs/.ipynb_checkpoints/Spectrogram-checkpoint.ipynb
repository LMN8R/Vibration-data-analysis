```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
```

## Good Horn

```python
data = pd.read_csv('Horn_vibration data Good horn.csv')
```

```python
time = np.array(data['Time(s)'])
amplitudes = []

for i in range(1, 13):
    amplitude = np.array(data['Amplitude(m/s2)%s' %(i)])
    amplitudes.append(amplitude)
```

```python
fig, axs = plt.subplots(2, 6)
fig.set_figheight(20)
fig.set_figwidth(20)
row, col = 0, 0
fig.suptitle('Spectrogram of Vibration Signals')

# Sampling frequency
fs = 1 / (time[1] - time[0])

for amplitude in amplitudes:
    
    # Compute the spectrogram
    frequencies, times, Sxx = signal.spectrogram(amplitude, fs)
    
    # Plot the spectrogram
    axs[row, col].pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='auto')
    # axs[row, col].ylabel('Frequency [Hz]')
    # axs[row, col].xlabel('Time [sec]')
    # axs[row, col].colorbar(label='Power/Frequency (dB/Hz)')
    col = (col + 1) % 6
    if (col == 5):
        row = (row + 1) % 2
```
