```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
```

```python
# Function to apply Fourier Transform and generate FFT graph
def generate_fft_graph(time, amplitude, title, plot=False):
    n = len(time)
    dt = time.iloc[1] - time.iloc[0]  # Time step
    freq = np.fft.fftfreq(n, dt)
    amplitude_fft = np.fft.fft(amplitude)
    
    # Plot the FFT graph (optional)
    if plot:
        plt.figure(figsize=(10, 6))
        plt.plot(freq, np.abs(amplitude_fft))
        plt.title(f'FFT Graph - {title}')
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Amplitude')
        plt.show()
    
    return freq, np.abs(amplitude_fft)
```

```python
# Load the data from CSV
centermass_data = pd.read_csv('centermassdamage.csv')
diaphragmdamage_data = pd.read_csv('diaphragmdamage.csv')
resonatordamage_data = pd.read_csv('resonatordamage.csv')
goodhorn_data = pd.read_csv('goodhorn.csv')
resonatordamage_data.head(10)
```

```python
goodhorn_data.head(10)
```

```python
# Extract time and amplitude columns
time_centermass = centermass_data['Time (sec)']
time_diaphragmdamage = diaphragmdamage_data['Time (s)']
time_resonatordamage = resonatordamage_data['Time(s)']
time_goodhorn = goodhorn_data['Time(s)']

amplitude_columns_centermass = centermass_data.columns[1:]
amplitude_columns_diaphragmdamage = diaphragmdamage_data.columns[1:]
amplitude_columns_resonatordamage = resonatordamage_data.columns[1:]
amplitude_columns_goodhorn = goodhorn_data.columns[1:]
```

```python
# Generate FFT graphs for each amplitude column
for column in amplitude_columns_centermass:
    freq, fft_data = generate_fft_graph(time_centermass, centermass_data[column], 'Center Mass Data', plot=True)

for column in amplitude_columns_diaphragmdamage:
    freq, fft_data = generate_fft_graph(time_diaphragmdamage, diaphragmdamage_data[column], 'Diaphragm Damage Data', plot=True)

for column in amplitude_columns_resonatordamage:
    freq, fft_data = generate_fft_graph(time_resonatordamage, resonatordamage_data[column], 'Resonator Damage Data', plot=True)

for column in amplitude_columns_goodhorn:
    freq, fft_data = generate_fft_graph(time_goodhorn, goodhorn_data[column], 'Good Horn Data', plot=True)
```
