```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import cwt, morlet
```

## Good Horn

```python
data = pd.read_csv('Horn_vibration data Good horn.csv')
```

```python
time = np.array(data['Time(s)'])
amplitudes = []

for i in range(1, 13):
    amplitude = np.array(data['Amplitude(m/s2)%s' %(i)])
    amplitudes.append(amplitude)
```

```python
fig, axs = plt.subplots(2, 6)
fig.set_figheight(20)
fig.set_figwidth(20)
row, col = 0, 0
fig.suptitle('Morlet Wavelet Transform of Vibration Signal')

# Parameters
sampling_rate = 1 / (time[1] - time[0])

# range of frequencies - can be adjusted later
wavelet_freqs = np.arange(1, 50, 1) 

for amplitude in amplitudes:
    
    # Morlet wavelet transform
    transformed_data = cwt(amplitude, morlet, wavelet_freqs, sampling_rate)
    
    # Plot the results
    axs[row, col].imshow(np.abs(transformed_data), aspect='auto', extent=[time.min(), time.max(), wavelet_freqs.min(), wavelet_freqs.max()], cmap='jet')
    # axs[row, col].ylabel('Frequency [Hz]')
    # axs[row, col].xlabel('Time [sec]')
    # axs[row, col].colorbar(label='Power/Frequency (dB/Hz)')
    col = (col + 1) % 6
    if (col == 5):
        row = (row + 1) % 2
```
